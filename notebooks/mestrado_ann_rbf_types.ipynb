```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%config InlineBackend.figure_format = 'svg'
params = {'figure.figsize': [5, 5],
'axes.labelsize': 20,
'axes.titlesize':24,
'font.size': 14,
'legend.fontsize': 14,
'xtick.labelsize': 18,
'ytick.labelsize': 18
}
plt.rcParams.update(params)
```

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']
```

```python
eps = 1
```

```python
plt.figure(figsize=(5, 5))
r = np.linspace(-4, 4, 1000)
y1 = np.exp(-(eps * r) ** 2)
y2 = np.sqrt(1 + (eps * r) ** 2)
plt.plot(r, y1, LineWidth=5, label=r'$\phi\ (r) = exp\left(-\left(\epsilon r\right)^{2}\right)$')
plt.plot(r, y2, LineWidth=5, label=r'$\phi\ (r) = \sqrt{1 + \left(\epsilon r\right)^{2}}$')
plt.legend()
plt.grid(True)
plt.xlim(-4, 4)
plt.ylim(-1, 4)
plt.xticks([-4, -2, 0, 2, 4])
plt.yticks([-1, 0, 1, 2, 3, 4])
plt.xlabel('r')
plt.ylabel(r'$\phi\ (r)$')


plt.savefig('../figs/ann_rbf_1.pdf', format='pdf', dpi=300, bbox_inches='tight', transparent=True)
plt.show()
```

```python
plt.figure(figsize=(5, 5))
r = np.linspace(-4, 4, 1000)
y3 = 1 / (1 + eps * r) ** 2
plt.plot(r, y3, LineWidth=5, label=r'$\phi\ (r) = \dfrac{1}{1 + \left(\epsilon r\right)^{2}}$')
plt.legend()
plt.grid(True)
plt.xlim(-4, 4)
plt.ylim(0, 120000)
plt.xticks([-4, -2, 0, 2, 4])
plt.yticks([0, 30000, 60000, 90000, 120000])
plt.xlabel('r')
plt.ylabel(r'$\phi\ (r)$')


plt.savefig('../figs/ann_rbf_2.pdf', format='pdf', bbox_inches='tight', transparent=True)
plt.show()
```

```python
plt.figure(figsize=(5, 5))
r = np.linspace(-4, 4, 1000)
y4 = 1 / np.sqrt((1 + eps * r) ** 2)
plt.plot(r, y4, LineWidth=5, label=r'$\phi\ (r) = \dfrac{1}{\sqrt{1 + \left(\epsilon r\right)^{2}}}$')
plt.legend()
plt.grid(True)
plt.xlim(-4, 4)
plt.ylim(0, 500)
plt.xticks([-4, -2, 0, 2, 4])
plt.yticks([0, 100, 200, 300, 400, 500])
plt.xlabel('r')
plt.ylabel(r'$\phi\ (r)$')


plt.savefig('../figs/ann_rbf_3.pdf', format='pdf', bbox_inches='tight', transparent=True)
plt.show()
```

```python
plt.figure(figsize=(5, 5))
r = np.linspace(-10, 10, 1000)
k1 = np.arange(1, 7, 2)
k2 = np.arange(2, 7, 2)
    
for z in range(1, 6):
    if z % 2 != 0:
        y = r ** z
    else:
        y = r ** z * np.log(r)
        
    plt.plot(r, y, LineWidth=5, alpha=0.85, label='k = %d' % z)

plt.legend()
plt.grid(True)
plt.xlim(-10, 10)
plt.ylim(-100000, 100000)
plt.xticks([-10, -5, 0, 5, 10])
plt.yticks([-100000, -50000, 0, 50000, 100000])
plt.xlabel('r')
plt.ylabel(r'$\phi\ (r)$')


plt.savefig('../figs/ann_rbf_4.pdf', format='pdf', bbox_inches='tight', transparent=True)
plt.show()
```
